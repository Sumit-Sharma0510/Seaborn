```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmri =  sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
sns.set_style("darkgrid")
sns.relplot(x = "timepoint", y= "signal",kind = "line",data = fmri)
```

```python
sns.relplot(x = "timepoint", y= "signal",kind = "line",data = fmri, ci = False)
```

```python
sns.relplot(x = "timepoint", y= "signal",kind = "line",hue = "event",data = fmri)
```

```python
sns.relplot(x = "timepoint", y= "signal",hue="region" , style = "event",kind = "line",data = fmri, ci = False)
```

```python
sns.relplot(x = "timepoint", y= "signal",hue="region" , style = "event",kind = "line",markers= True,dashes=False ,data = fmri)
```

```python
sns.relplot(x = "timepoint", y= "signal",hue="region" , style = "event",kind = "line",markers= True,data = fmri, ci = 68)
```
